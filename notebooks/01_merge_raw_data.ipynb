# Merge Data

Merge raw data from different periods and sources

```python
import sys
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Add src to path
sys.path.insert(0, str(Path('../src').resolve()))
from merge_data import merge_microclimate_excel_data, merge_microclimate_txt_data, merge_power_data, merge_weather_data, display_summary

print("Data Merging Stage")
print("Combines raw microclimate, weather, and power data from multiple files")
```

## Configuration

```python
# Paths
DATA_DIR = Path('../raw_data')
MICROCLIMATE_RAW = DATA_DIR / 'Raw_Microclimate_download'
WEATHER_RAW = DATA_DIR / 'Raw_Weather_download'
POWER_RAW = DATA_DIR / 'Raw_Power_download'

# Use pipeline output directory if running from main pipeline, otherwise use default
import os
if 'PIPELINE_OUTPUT_ROOT' in os.environ:
    OUTPUT_DIR = Path(os.environ['PIPELINE_OUTPUT_ROOT']) / 'merged'
else:
    OUTPUT_DIR = DATA_DIR / 'merged'

# Validate input directories
if not MICROCLIMATE_RAW.exists():
    print(f"WARNING: Microclimate data directory not found: {MICROCLIMATE_RAW}")
if not WEATHER_RAW.exists():
    print(f"WARNING: Weather data directory not found: {WEATHER_RAW}")
if not POWER_RAW.exists():
    print(f"WARNING: Power data directory not found: {POWER_RAW}")

# Create output directories
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
(OUTPUT_DIR / 'microclimate').mkdir(exist_ok=True)
(OUTPUT_DIR / 'weather').mkdir(exist_ok=True)
(OUTPUT_DIR / 'power').mkdir(exist_ok=True)


print(f"Configuration:")
print(f"  Output: {OUTPUT_DIR}")
print(f"  Microclimate input: {MICROCLIMATE_RAW}")
print(f"  Power input: {POWER_RAW}")
print(f"  Weather input: {WEATHER_RAW}")
```

## Merge Microclimate Data

```python
# Merge Excel microclimate data (Time1-9.xlsx)
merge_microclimate_excel_data(MICROCLIMATE_RAW, OUTPUT_DIR / 'microclimate')

# Merge txt microclimate data - Library Green Roof
merge_microclimate_txt_data(MICROCLIMATE_RAW / 'Lib_GR', OUTPUT_DIR / 'microclimate', 'Lib_GR')

# Merge txt microclimate data - PV Lift3
merge_microclimate_txt_data(MICROCLIMATE_RAW / 'PV_Lift3', OUTPUT_DIR / 'microclimate', 'PV_Lift3')

print("\nMicroclimate data merge complete")
```

## Merge Weather Data

```python
merge_weather_data(WEATHER_RAW, OUTPUT_DIR / 'weather')
```

## Merge Power Data

```python
merge_power_data(POWER_RAW, OUTPUT_DIR / 'power')
```

## Summary

```python
display_summary(OUTPUT_DIR)
```
