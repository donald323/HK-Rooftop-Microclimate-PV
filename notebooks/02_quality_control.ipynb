# Quality Control on Merged Data

Apply comprehensive quality control to merged data files (microclimate, power, and weather):
- Constant value detection (automated failure detection)
- Sensor failure periods (documented failures)
- Boundary checks

```python
print("Quality Control Stage")
print("Applying comprehensive quality control to merged data")
print()

import sys
from pathlib import Path
import warnings
import pandas as pd
warnings.filterwarnings('ignore')

# Add src to path
sys.path.insert(0, str(Path('../src').resolve()))
from quality_control import QualityController
```

## Configuration

```python
# Paths
DATA_DIR = Path('../data')

# Use pipeline output directory if running from main pipeline
import os
if 'PIPELINE_OUTPUT_ROOT' in os.environ:
    OUTPUT_ROOT = Path(os.environ['PIPELINE_OUTPUT_ROOT'])
    MERGED_DIR = OUTPUT_ROOT / 'merged'
    QC_DIR = OUTPUT_ROOT / 'quality_controlled'
else:
    MERGED_DIR = DATA_DIR / 'merged'
    QC_DIR = DATA_DIR / 'quality_controlled'

# Validate input directory
if not MERGED_DIR.exists():
    print(f"WARNING: Merged data directory not found: {MERGED_DIR}")

# Create output directories
QC_DIR.mkdir(parents=True, exist_ok=True)
(QC_DIR / 'microclimate').mkdir(exist_ok=True)
(QC_DIR / 'power').mkdir(exist_ok=True)
(QC_DIR / 'weather').mkdir(exist_ok=True)

# Initialize quality controllers
microclimate_qc = QualityController(data_type='microclimate')
power_qc = QualityController(data_type='power')
weather_qc = QualityController(data_type='weather')

print(f"Configuration:")
print(f"  Input: {MERGED_DIR}")
print(f"  Output: {QC_DIR}")
print(f"  Processing: Constant Values → Failures → Boundaries")
```

# Microclimate Data Quality Control

```python
print("Processing microclimate data...")
microclimate_files = list((MERGED_DIR / 'microclimate').glob('*.csv'))

if not microclimate_files:
    print("No microclimate files found in merged directory")
else:
    for file in sorted(microclimate_files):
        output_file = QC_DIR / 'microclimate' / file.name
        print(f"  {file.name}")
        qc_stats = microclimate_qc.apply_to_file(file, output_file)

    print(f"Completed: {len(microclimate_files)} microclimate files")
```

# Power Data Quality Control

```python
print("Processing power data...")
power_files = list((MERGED_DIR / 'power').glob('*.csv'))

if not power_files:
    print("No power files found in merged directory")
else:
    for file in sorted(power_files):
        output_file = QC_DIR / 'power' / file.name
        print(f"  {file.name}")
        qc_stats = power_qc.apply_to_file(file, output_file)

    print(f"Completed: {len(power_files)} power files")
```

# Weather Data Quality Control

```python
print("Processing weather data...")
weather_files = list((MERGED_DIR / 'weather').glob('*.csv'))

if not weather_files:
    print("No weather files found in merged directory")
else:
    for file in sorted(weather_files):
        output_file = QC_DIR / 'weather' / file.name
        print(f"  {file.name}")
        qc_stats = weather_qc.apply_to_file(file, output_file)

    print(f"Completed: {len(weather_files)} weather files")
```
